```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to generate cone surface
def plot_cone(radius, height, n_points=50):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Create the cone surface
    z = np.linspace(0, height, n_points)
    theta = np.linspace(0, 2 * np.pi, n_points)
    Z, Theta = np.meshgrid(z, theta)
    X = (height - Z) * np.cos(Theta) * radius / height
    Y = (height - Z) * np.sin(Theta) * radius / height

    # Plot the cone surface
    ax.plot_surface(X, Y, Z, color='orange', alpha=0.6)

    # Create the base (circle at the bottom)
    theta_circle = np.linspace(0, 2 * np.pi, n_points)
    X_base = radius * np.cos(theta_circle)
    Y_base = radius * np.sin(theta_circle)
    Z_base = np.zeros_like(X_base)

    # Plot the base of the cone
    ax.plot(X_base, Y_base, Z_base, color='blue')

    # Set labels
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")

    plt.show()

# Example: Plot cone with radius 5 and height 10
plot_cone(radius=5, height=10)
```
